```python
import os
import random
from PIL import Image, ImageDraw, ImageFont
another_100_sentences = [
    "She ran through the field chasing after butterflies",
    "The old clock in the hallway chimed loudly",
    "Beneath the stars they shared stories of their childhood",
    "A gentle rain began to fall on the quiet town",
    "He opened the box to reveal a mysterious object",
    "The garden was filled with the scent of blooming flowers",
    "Waves crashed against the shore with relentless force",
    "The children gathered stones to build a small fort",
    "The fireflies danced in the warm summer night",
    "A soft breeze rustled the pages of the open book",
    "The mountain air was crisp and refreshing at dawn",
    "She painted the walls with bright cheerful colors",
    "The cat slept peacefully on the sunny windowsill",
    "The river sparkled as sunlight kissed its surface",
    "Leaves fluttered down from the ancient oak tree",
    "He listened carefully to the whispers of the wind",
    "The old library smelled of leather and dust",
    "They hiked through the forest following a narrow path",
    "The aroma of freshly baked cookies filled the house",
    "Clouds gathered quickly signaling an approaching storm",
    "The boy balanced carefully on the edge of the dock",
    "She folded the letter and tucked it into her pocket",
    "The stars above twinkled with a mysterious glow",
    "A gentle hum filled the room as the machine whirred",
    "The snow fell softly covering the ground in white",
    "He found a forgotten photograph in the attic",
    "The wind carried the sound of distant laughter",
    "The dog chased after the ball across the open field",
    "She wrapped the warm scarf tightly around her neck",
    "The sun dipped below the horizon painting the sky",
    "A lone tree stood tall in the middle of the meadow",
    "The children’s voices echoed through the empty halls",
    "He traced patterns in the sand with his fingertips",
    "The old bridge swayed slightly in the strong wind",
    "She watched the clouds drift lazily across the sky",
    "The fireplace crackled sending sparks into the air",
    "The garden gate creaked open revealing a secret path",
    "A flock of birds took flight at the break of dawn",
    "The mountain peak was covered in a blanket of snow",
    "He whispered a wish as he blew out the candle",
    "The river’s song was soothing to her tired mind",
    "The children danced joyfully around the maypole",
    "She collected shells scattered along the sandy shore",
    "The clock’s hands moved steadily toward midnight",
    "The scent of pine trees filled the crisp morning air",
    "He climbed higher until he reached the rocky ledge",
    "The moon cast a silver glow over the quiet lake",
    "She hummed a tune while knitting a colorful sweater",
    "The garden bloomed with roses in shades of red and pink",
    "A cat’s eyes gleamed mysteriously in the dark alley",
    "The waves lapped gently at the wooden pier",
    "He found comfort sitting beneath the ancient tree",
    "The sky turned brilliant shades of orange and pink",
    "The children’s laughter bubbled like a mountain stream",
    "She pressed petals carefully into the pages of her book",
    "The wind carried the scent of fresh rain on dry earth",
    "The boat rocked gently as it floated on the calm sea",
    "He wrote stories that transported readers to faraway lands",
    "The forest was alive with the chirping of crickets",
    "The kitten pawed at a ball of tangled yarn",
    "She watched as the first snowflakes began to fall",
    "The old house stood silent beneath a canopy of stars",
    "The morning sun warmed the dew on the grass",
    "He folded paper airplanes and sent them soaring",
    "The children built castles out of soft golden sand",
    "A gentle ripple spread across the surface of the pond",
    "The leaves rustled softly under the feet of a passing deer",
    "She traced her fingers over the cool stone wall",
    "The bakery filled the street with the smell of fresh bread",
    "The lantern flickered casting shadows on the cobblestones",
    "The mountain trail wound through dense forests and open meadows",
    "He found a lost puppy hiding under a bush",
    "The sky darkened as storm clouds rolled in",
    "She smiled as the cool breeze tousled her hair",
    "The garden was alive with buzzing bees and colorful butterflies",
    "He listened to the distant sound of church bells ringing",
    "The river carved its way through the deep canyon",
    "The children’s footsteps pitter pattered on the wooden floor",
    "She planted seeds that would grow into tall sunflowers",
    "The old piano played a haunting melody in the empty room",
    "A swarm of fireflies lit up the summer night",
    "The cat chased a shadow flickering on the wall",
    "He held the map tightly as they ventured into the unknown",
    "The snow crunched underfoot as they walked through the woods",
    "She wrapped herself in a blanket by the glowing campfire",
    "The leaves shimmered with morning frost",
    "The breeze whispered secrets through the tall grass",
    "He sat quietly watching the sun set over the hills",
    "The children’s voices rose in song around the campfire",
    "The scent of lavender filled the warm summer air",
    "She found a feather that glowed faintly in the moonlight",
    "The waves whispered stories of distant lands",
    "He carved a figure from a piece of driftwood",
    "The garden path was lined with blooming tulips",
    "A gentle rain kissed the petals of the flowers",
    "The old clock tower tolled the hour slowly",
    "She danced barefoot beneath the light of the stars",
    "The breeze carried the laughter of children playing nearby",
    "He watched as the clouds rolled across the sky",
    "The kitten purred softly as it curled in her lap",
    "The lantern light flickered in the quiet night",
    "She pressed a flower into the pages of her journal",
    "The river flowed quietly beneath the wooden bridge",
    "He dreamed of adventures in faraway places",
    "The forest was peaceful except for the rustling leaves",
    "The children splashed joyfully in the shallow stream",
    "She traced the patterns of frost on the windowpane",
    "The stars shone brightly above the sleeping village"
]








# === SETTINGS ===
font_path = r"little_kids_handwriting/LittleKidsHandwriting-Regular.otf"
output_size = (900, 200)
base_font_size = 32

# === SWAPPABLE LETTERS ===
swappable_letters_all = {
    'b': ('d', 0.0), 'd': ('b', 0.0),
    'p': ('q', 0.0), 'q': ('p', 0.0),
    'm': ('w', 0.0), 'w': ('m', 0.0),
    'n': ('u', 0.0), 'u': ('n', 0.0),
    'i': ('l', 0.0), 'l': ('i', 0.0),
    's': ('z', 0.0), 'z': ('s', 0.0),
    'o': ('u', 0.0)
}


def create_image(sentence, base_font, distort=False, swap_chance=0.5):
    img = Image.new("RGB", output_size, color="white")
    draw = ImageDraw.Draw(img)

    if not distort:
        x, y = 10, 30
        max_width = output_size[0] - 20
        line_spacing = base_font_size + 10
        words = sentence.split()
        line = ""

        for word in words:
            test_line = line + word + " "
            bbox = base_font.getbbox(test_line)
            w = bbox[2] - bbox[0]
            if w < max_width:
                line = test_line
            else:
                draw.text((x, y), line, font=base_font, fill="black")
                y += line_spacing
                line = word + " "

        if line:
            draw.text((x, y), line, font=base_font, fill="black")

    else:
        swappable_letters = {k: (v[0], swap_chance) for k, v in swappable_letters_all.items()}
        sentence_words = sentence.split()

        x, y = 10, 30
        for word in sentence_words:
            distorted_word = ""
            for char in word:
                if char in swappable_letters:
                    replacement, char_swap_chance = swappable_letters[char]
                    if random.random() < char_swap_chance:
                        distorted_word += replacement
                    else:
                        distorted_word += char
                else:
                    distorted_word += char

            # Create the full word image (opaque white background to avoid blur)
            word_img = Image.new("RGB", (300, 50), "white")
            word_draw = ImageDraw.Draw(word_img)

            temp_x = 0
            for char in distorted_word:
                font_size = base_font_size + random.randint(-2, 2)
                try:
                    font = ImageFont.truetype(font_path, font_size)
                except:
                    font = base_font

                char_width = font.getbbox(char)[2] - font.getbbox(char)[0]
                offset_y = random.randint(-2, 2)

                # Always solid black ink
                word_draw.text((temp_x, offset_y), char, font=font, fill="black")
                temp_x += char_width + random.randint(1, 4)

            word_width = temp_x

            if x + word_width > output_size[0] - 10:
                x = 10
                y += base_font_size + 15

            img.paste(word_img, (x, y))
            x += word_width + 10

    return img


# === LOAD FONT ===
try:
    base_font = ImageFont.truetype(font_path, base_font_size)
except IOError:
    raise Exception(f"Font not found: {font_path}")


def get_next_index(directory):
    files = [f for f in os.listdir(directory) if f.endswith(".png")]
    indices = [int(os.path.splitext(f)[0]) for f in files if os.path.splitext(f)[0].isdigit()]
    return max(indices, default=0) + 1

# === GENERATE IMAGES FOR EACH SWAP LEVEL ===
swap_levels = {
    "25": 0.25,
    "50": 0.50,
    "75": 0.75
}

normal_dir = "generated_data/normal"
os.makedirs(normal_dir, exist_ok=True)

for level in swap_levels:
    os.makedirs(f"generated_data/dyslexic/{level}", exist_ok=True)

# Ensure unique filenames across all folders
all_dirs = [normal_dir] + [f"generated_data/dyslexic/{level}" for level in swap_levels]
start_index = max(get_next_index(d) for d in all_dirs)

for i, sentence in enumerate(sentences, start=start_index):
    filename = f"{i:03}.png"

    # Save normal image
    normal_img = create_image(sentence, base_font, distort=False)
    normal_img.save(os.path.join(normal_dir, filename))

    # Save dyslexic images for each distortion level
    for level, chance in swap_levels.items():
        dys_img = create_image(sentence, base_font, distort=True, swap_chance=chance)
        dys_img.save(os.path.join(f"generated_data/dyslexic/{level}", filename))

print("✅ All normal and dyslexic images generated with rich black ink and non-overlapping filenames.")

```
